```python

x = []
y = []

files = os.listdir("pelak")

for image_name in files:
    file_path = os.path.join("pelak", image_name)
    im = cv2.imread(file_path)
    if im is None:
        continue
    im2 = cv2.resize(im, (32, 32))
    im3 = cv2.cvtColor(im2, cv2.COLOR_BGR2GRAY)
    im4 = im3.flatten()
    x.append(im4)
    try:
        label = int(image_name.split(".")[0])
    except ValueError:
        continue
    y.append(label)

x = np.array(x)
y = np.array(y)
```
